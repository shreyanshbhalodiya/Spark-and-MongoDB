## 1.Create array of names and publish name and length of each name.

```python
# Create data in Plain RDD
from pyspark import SparkContext
sc = SparkContext.getOrCreate()

#--Spark version 2.4 and python 3.8 issue
#array= sc.parallelize(    range(10))
array= sc.parallelize(["Greta","Maria","Gracia","Laura"])
# Print data in RDD array
#array=str(array)
#print (array)
array.collect()
# Extract lenth of each element in array
x = sc.parallelize(["Greta","Maria","Gracia","Laura"])
y = x.map(lambda x: (x,len(x)))
y.collect()
#type(array)
```

# a.List no. of shows per year.

```python
data = sc.textFile('C:\\sparkinstall\\spark-2.2.0-bin-hadoop2.7\\bin\DATASET_FILES\\ocp_data.tsv')
```

```python
data.take(5)
```

```python
# Splitting the RDD :
data = data.map(lambda line:line.split('\t'))
```

```python
data.take(5)
```

```python
# Getting the columns RDD :
columns = data.first()
```

```python
# Removing columns so that we will get all the rows records fro dataset:
rows = data.filter(lambda l: l!=columns)
```

```python
years1 = rows.map(lambda l: (l[0]))
```

```python
print("Year and the corresponding no of shows (in dictionary form of key(YEAR) and value(No of shows) ) : ")
years1.countByValue()
```

## b.List no. of shows per year for every occupation¶

```python
years_occup = rows.map(lambda l: (l[0], l[1]))
```

```python
print("The Year, Occupation and their corresponding no. of shows: \n (NOTE: in the format of {(YEAR, Occupation): no_of_shows}) below result: ")
years_occup.countByValue()
```

```python
print("The Year, Occupation and their corresponding no. of shows: \n (NOTE: Year, Occupation and count is the no_of_shows: result below: )")
sorted(df.groupBy(['YEAR', 'GoogleKnowlege_Occupation']).count().collect())
```

## c.Find top occupation for every year.

```python
from collections import Counter
top_occup = rows.map(lambda x: (x[0],x[1])).groupByKey().mapValues(list).map(lambda x: (x[0],Counter(x[1]).most_common(1)))
top_occup.sortByKey().collect()
```

##  d. Which guest have attended maximum shows¶

guest_shows = rows.map(lambda l: (l[4],1))

```python
guest_shows2 = guest_shows.reduceByKey(lambda a,b:a+b)
```

```python
print("The Guest name who attended max no. of shows: ")
guest_shows2.sortBy(ascending=False, keyfunc=lambda guest_shows: guest_shows[1]).first()
```

## e.List occupation under each group

```python
group_occup = rows.map(lambda l: (l[3], l[1]))
```

```python
print("The Groups followed by the Occupation list is as follows:\nNOTE: in the format of (Group, Occupation) as below result:\n")
group_occup.sortBy(lambda l: sorted(l)).distinct().collect()
```

## f. How many show per group of year and occupation

```python
group_year_occup = rows.map(lambda l: (l[3], l[0], l[1]))
```

```python
print("The Group, Year, Occupation and their corresponding no. of shows: \n (NOTE: in the format of {(Group, YEAR, Occupation): no_of_shows}) below result:\n")
group_year_occup.countByValue()
```
